```python
import re
```

```python
with open('aoc7.txt', 'r') as file:
    f = file.readlines()
    
# f
list_1 = [i.strip('\n') for i in f]
```

```python
def contents(i):
    container = re.split(r' bags?,? contain ?', i)[0]
    inside = re.split(r'[,.]', re.split(r' bags?,? contain ?', i)[1])[:-1]
    inside = [re.sub('bags?', '', re.sub('\d+\s+?', '', i)).strip() for i in inside]
    return container, inside

def bag_checker(bag_type, list_1, count):
    for i in (list_1):
        container, inside = contents(i)
        if bag_type in inside:
            print(i)
            count += 1
            container_list.append(container)
    print(count)
    if bag_type in container_list:
        container_list.remove(bag_type)
    return(count, container_list)
```

```python
# Part 1
print('Part 1')
count = 0
count_updated = 0
container_list = []
iteration = 1
while iteration == 1 or (iteration != 1 and len(container_list) != 0):
    if iteration == 1:
        count_updated, container_list = bag_checker('shiny gold', list_1, count)
        print(f'Iteration 1 Count: {count_updated}')
    else:
#         pass
#         container_list = []
        for i in container_list:
            count_updated, container_list = bag_checker(i, list_1, count_updated)
            iteration += 1
            print(f'Iteration {iteration} Count: {count_updated}')
        
# for i in (list_1):
#     container, inside = contents(i)
#     if 'shiny gold' in inside:
#         print(i)
#         count += 1
#         container_list.append(container)
#     print(container)
#     print(inside)
print(count_updated, container_list)
```
